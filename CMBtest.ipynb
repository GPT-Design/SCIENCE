## **Section 2: Data Loading (Planck CMB Maps)**
(Markdown cell)

Clearly load and visualize the Planck CMB data explicitly from your Google Drive. Verify file loading and inspect data quality explicitly.

```python
# Explicit data directory (Adjust if needed)
data_dir = "/content/drive/MyDrive/AI/3TE/CMB/DATA"

# Explicit Planck file names clearly stated
cmb_files = {
    "030 GHz": "npipe6v20_030_map.fits",
    "044 GHz": "npipe6v20_044_map.fits",
    "070 GHz": "npipe6v20_070_map.fits",
    "857 GHz": "npipe6v20_857_map.fits"
}

# Load CMB maps explicitly into a dictionary
cmb_maps = {}

for freq, file in cmb_files.items():
    file_path = os.path.join(data_dir, file)
    print(f"Loading {freq} data explicitly from: {file}")

    try:
        cmb_maps[freq] = hp.read_map(file_path, verbose=False)
        print(f"{freq} loaded explicitly: shape={cmb_maps[freq].shape}")
    except Exception as e:
        print(f"Error explicitly loading {freq}: {e}")
```

## **Section 3: Visualizing Planck CMB Maps**
(Markdown cell)

Explicitly verify the loaded Planck maps visually, clearly confirming data integrity.

```python
# Visualization settings clearly defined
for freq, cmb_map in cmb_maps.items():
    plt.figure(figsize=(12, 6))
    hp.mollview(cmb_map, title=f"Planck CMB Intensity Map ({freq})", cmap='coolwarm', unit='K', norm='hist')
    plt.show()
```

## **Section 4: Computing Observed CMB Angular Power Spectra**
(Markdown cell)

Compute the angular power spectra explicitly using standard spherical harmonic transforms (Healpy anafast method clearly used).

```python
# Compute angular power spectra explicitly (Cl values)
cmb_spectra = {}

for freq, cmb_map in cmb_maps.items():
    print(f"Explicitly computing Cl spectrum for {freq}...")
    cmb_spectra[freq] = hp.anafast(cmb_map, lmax=2500)  # lmax adjustable explicitly

    # Clearly plot spectrum
    plt.figure(figsize=(10, 6))
    plt.plot(cmb_spectra[freq], label=f'{freq}')
    plt.xlabel('Multipole moment (l)')
    plt.ylabel(r'$C_\ell$')
    plt.title(f'Planck CMB Angular Power Spectrum ({freq})')
    plt.legend()
    plt.grid(True)
    plt.show()
```

## **Section 6: Rigorous Statistical Validation (Monte Carlo & Bootstrap)**

```python
# Rigorous Monte Carlo Statistical Test
def monte_carlo_validation(observed, predicted, n_sim=1000):
    np.random.seed(42)  # Explicit reproducibility
    observed_mean = np.mean(observed)
    predicted_mean = np.mean(predicted)

    residual_mean = observed_mean - predicted_mean
    combined_std = np.std(observed - predicted)

    simulated_residuals = np.random.normal(0, combined_std, n_sim)
    p_value = np.mean(np.abs(simulated_residuals) >= np.abs(residual_mean))

    return residual_mean, combined_std, p_value

# Execute robust statistical validation
residual_mean, residual_std, p_val = monte_carlo_validation(observed_spectrum, predicted_spectrum)

# Explicitly report statistical results clearly:
print(f"Mean residual: {residual_mean:.2e}")
print(f"Residual standard deviation: {residual_std:.2e}")
print(f"P-value of observed difference explicitly: {p_val:.3f}")

# Interpret explicitly:
if p_val < 0.05:
    print("Explicit conclusion: Significant difference (p < 0.05). 3T_E may require adjustment.")
else:
    print("Explicit conclusion: No statistically significant difference. 3T_E predictions align robustly with Planck observations.")
```

```python
# Bootstrap Confidence Interval Calculation explicitly:
def bootstrap_confidence_interval(data, n_bootstrap=1000, ci=95):
    np.random.seed(42)
    boot_means = np.array([
        np.mean(np.random.choice(data, size=len(data), replace=True))
        for _ in range(n_bootstrap)
    ])
    lower_bound = np.percentile(boot_means, (100 - ci) / 2)
    upper_bound = np.percentile(boot_means, 100 - (100 - ci) / 2)
    return lower_bound, upper_bound

# Explicitly calculate confidence intervals for residuals:
ci_lower, ci_upper = bootstrap_confidence_interval(residual)

# Explicitly report CI results:
print(f"95% CI for mean residual: [{ci_lower:.2e}, {ci_upper:.2e}]")

# Visualize explicitly:
plt.figure(figsize=(10, 6))
plt.hist(residual, bins=50, color='orange', alpha=0.7, edgecolor='k')
plt.axvline(ci_lower, color='red', linestyle='--', label='95% CI Lower')
plt.axvline(ci_upper, color='blue', linestyle='--', label='95% CI Upper')
plt.title('Bootstrap Confidence Interval of Residual')
plt.xlabel('Residual Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
